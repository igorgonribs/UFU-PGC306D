# **Bibliotecas**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

import requests
```

**Carrega a imagem**

```python
# URL da imagem (exemplo da Lena em grayscale)
url = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQRi0WuTBazwcDDGwLnErgLw00hrUjnJfT3jA&s" # imagem da Lenna

# Faz o download
resp = requests.get(url, stream=True).content

# Converte para array numpy
arr = np.asarray(bytearray(resp), dtype=np.uint8)

# Decodifica como imagem
img = cv2.imdecode(arr, cv2.IMREAD_GRAYSCALE)
img
```

**Aplicação de Sobel (Horizontal)**

```python
# máscara sobel (horizontal)

h = np.array([[-1, -2, -1],
              [ 0,  0 , 0],
              [ 1,  2,  1]
              ])

# aplicação do filtro com o OpenCV
mascara_h = cv2.filter2D(img, cv2.CV_64F, h)

# plot do resultado
fig, axs = plt.subplots(1, 2, figsize=(9, 5))

# imagem original
axs[0].imshow(img, cmap='gray')
axs[0].set_title('Imagem original')

# imagem filtrada
axs[1].imshow(mascara_h, cmap='gray')
axs[1].set_title('Sobel Horizontal')
```
